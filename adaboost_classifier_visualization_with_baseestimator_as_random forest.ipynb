```python
import matplotlib.pyplot as plt
%matplotlib inline
import warnings
warnings.filterwarnings("ignore")
```

```python
# load iris dataset
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
# feature names
features_name = iris.feature_names
#target names
target_name = iris.target_names
```

```python
# load required library
import numpy as np
from sklearn.datasets import load_iris
from sklearn.ensemble import AdaBoostClassifier, RandomForestClassifier
from sklearn.tree import  plot_tree
```

```python
def plot_decision_trees(clf, iteration, feature_name, target_name):
    plt.figure(figsize=[7.0,7.0])
    plot_tree(clf, filled=True, feature_names=feature_name, class_names=target_name, node_ids=True)
    plt.show()
    print("Random Forest {}".format(iteration))
```

```python
base_estimator = RandomForestClassifier(n_estimators=5)
```

```python
model = AdaBoostClassifier(base_estimator=base_estimator, n_estimators=5)
model.fit(iris.data, iris.target)
```

```python
model.estimators_
```

```python
for i in range(len(model.estimators_)):
    for j in range(len(model.estimators_[i].estimators_)):
        plot_decision_trees(model.estimators_[i].estimators_[j], (j+1 * i+1), features_name, target_name)
```
